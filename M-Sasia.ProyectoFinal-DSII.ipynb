```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import json
import time

# Spotify API credentials
SPOTIPY_CLIENT_ID = 'e83bd9d42fb94d8a83fc0b287001f745'
SPOTIPY_CLIENT_SECRET = '996af9f2ffb9434eaa01c6cd253c1ce0'

# Set up Spotify API client
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID,
                                                           client_secret=SPOTIPY_CLIENT_SECRET))

def get_tracks_by_year(year):
    """Fetch tracks from the specified year using Spotify API."""
    results = sp.search(q=f'year:{year}', type='track', limit=1)
    tracks = results['tracks']['items']
    
    # Handle pagination
    while results['tracks']['next']:
        time.sleep(1)  # Avoid hitting rate limits
        results = sp.next(results['tracks'])
        tracks.extend(results['tracks']['items'])
    
    return tracks

def extract_track_info(track):
    """Extract relevant information from a track."""
    return {
        'id': track['id'],
        'name': track['name'],
        'artists': ', '.join(artist['name'] for artist in track['artists']),
        'duration_ms': track['duration_ms'],
        'release_date': track['album']['release_date'],
        'year': track['album']['release_date'][:4],
        'acousticness': track.get('acousticness', 0),
        'danceability': track.get('danceability', 0),
        'energy': track.get('energy', 0),
        'instrumentalness': track.get('instrumentalness', 0),
        'liveness': track.get('liveness', 0),
        'loudness': track.get('loudness', 0),
        'speechiness': track.get('speechiness', 0),
        'tempo': track.get('tempo', 0),
        'valence': track.get('valence', 0),
        'mode': track.get('mode', 0),
        'key': track.get('key', 0),
        'popularity': track.get('popularity', 0),
        'explicit': track['explicit'],
        'genre': ', '.join(sp.artist(track['artists'][0]['id'])['genres'])  # Simplified genre extraction
    }

def create_dataframe(years):
    """Create a DataFrame with track information for specified years."""
    all_tracks = []
    for year in years:
        print(f"Fetching tracks for year {year}...")
        tracks = get_tracks_by_year(year)
        for track in tracks:
            all_tracks.append(extract_track_info(track))
        time.sleep(5)  # To avoid hitting rate limits

    df = pd.DataFrame(all_tracks)
    return df

# Define the range of years
years = range(1950, 2024)

# Create DataFrame
df = create_dataframe(years)

# Save DataFrame to CSV
df.to_csv('spotify_tracks_1950_2023.csv', index=False)
print("Dataset saved to spotify_tracks_1950_2023.csv")
```
